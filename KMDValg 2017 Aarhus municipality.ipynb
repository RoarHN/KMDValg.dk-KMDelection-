```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome("/Users/RoarH/Python/chromedriver")
```

```python
partys=[] #List to store party of the named candidate
locations=[] #List to store location where named candidate received the votes
nameList=[] #List to store name of the candidate
votesList=[] #List to store number of votes for the named candidate
eligibleVotersAtLocation=[] #list to store the total votes eligible voters at a given location
```

```python
URL="https://www.kmdvalg.dk/KV/2017/K84713751.htm"
driver.get(URL)
content = driver.page_source
soup = BeautifulSoup(content)

links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))
del(links[:24])

fullLinks=[]
for link in links:
    link=link[:12]
    fullLink=(link)
    fullLinks.append(fullLink)

allLink=[]
allLinks=[]
partyLetters = ['A', 'B', 'C', 'D', 'E', 'F', 'I', 'K', 'L', 'N', 'O', 'P', 'R', 'T', 'V', 'Ø', 'Å']
for link in fullLinks:
    for party in partyLetters:
        allLink=('https://www.kmdvalg.dk/KV/2017/'+link+party+'.htm')
        allLinks.append(allLink)
```

```python
for URL in allLinks:
    driver.get(URL)
    content = driver.page_source
    soup = BeautifulSoup(content)

    names=soup.findAll('div', attrs={'class':'table-like-cell col-xs-7 col-sm-6 col-md-6 col-lg-8'})
    for name in names:
        nameList.append(name.text)
    URL=URL[:44]
    URL=URL[43:]
    nameList.append('Listestemmer'+' '+URL) 
    
    votess=soup.findAll('div', attrs={'class':'table-like-cell col-xs-5 col-sm-6 col-md-6 col-lg-4 text-right roboto-bold'})
    for votes in votess:
        votesList.append(votes.text)
    votes=soup.find('td', attrs={'class':'col-xs-6 roboto-bold text-right'}).findNext('td', attrs={'class':'col-xs-6 roboto-bold text-right'})
    votesList.append(votes.text)

    a=len(names)+1
    b=0
    while b<a:
        party=soup.find('span', attrs={'class':'content-block-sub-header'}).findNext('span', attrs={'class':'content-block-sub-header'})
        partys.append(party.text[1:7])
        location=soup.find("a").findNext("a").findNext("a").findNext("a").findNext("a").findNext("a").findNext("a")
        locations.append(location.text)
        b=b+1
```

```python
parentLinks=[] #List to store all parent links
for Links in fullLinks:
    parentLink=('https://www.kmdvalg.dk/KV/2017/'+Links+'.htm')
    parentLinks.append(parentLink)
for URL in parentLinks:
    driver.get(URL)
    content = driver.page_source
    soup = BeautifulSoup(content)
    eligible=soup.find('td').findNext('td')
    a=0
    while a<220:
        eligibleVotersAtLocation.append(eligible.text)
        a=a+1
print(len(partys), len(locations), len(nameList), len(votesList), len(eligibleVotersAtLocation))
```

```python
df = pd.DataFrame({'Party':partys,'Location':locations,'Name':nameList,'Votes':votesList, 'eligibleVoters':eligibleVotersAtLocation}) 
df.to_csv('Kommunalvalg 2017 Resultat.csv', index=False, encoding='utf-8')
```
